### User-defined Functions (UDFs)

- To print total model parameters

```python
def check_params(model, *args, **kwargs) -> dict:
    return {
        "total_trainable_params" : sum(p.numel() for p in model.parameters() if p.requires_grad),
        "total_nontrainable_params" : sum(p.numel() for p in model.parameters() if not p.requires_grad)
    }
```

- To get the pair of train and validation dataloaders

```python
data_transform = torchvision.transforms.Compose([
    torchvision.transforms.Resize((224, 224)),
    torchvision.transforms.ToTensor(),
    torchvision.transforms.Normalize((torch.tensor(33.3184)/255,), (torch.tensor(78.5675)/255,))
])

train_dataset = torchvision.datasets.MNIST(root=".", train=True, transform=data_transform, download=True).train_data.float()

def get_dataloaders(train_batch_size : int, val_batch_size : int, max_rows : int = 1000, *args, **kwargs) -> tuple:
    data_transform = torchvision.transforms.Compose([
        torchvision.transforms.Resize((224, 224)),
        torchvision.transforms.ToTensor(),
        torchvision.transforms.Normalize((torch.tensor(33.3184)/255,), (torch.tensor(78.5675)/255,))
    ])

    train_dataset = torchvision.datasets.MNIST(root=".", train=True, transform=data_transform)
    train_idx = torch.randperm(len(train_dataset))[:int(max_rows*.75)]
    train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=train_batch_size, \
                                               sampler=torch.utils.data.SubsetRandomSampler(train_idx))

    val_dataset = torchvision.datasets.MNIST(root=".", train=False, transform=data_transform)
    val_idx = torch.randperm(len(val_dataset))[:int(max_rows*.25)]
    val_loader = torch.utils.data.DataLoader(val_dataset, batch_size=val_batch_size, \
                                             sampler=torch.utils.data.SubsetRandomSampler(val_idx))

    return train_loader, val_loader
```

* To fit (training) the model

```python
def fit(
    model : torchvision.models,
    epoch : int,
    train_loader : torch.utils.data.DataLoader,
    val_loader : torch.utils.data.DataLoader,
    *args, **kwargs
) -> dict:

    TRAIN_LOSS, TRAIN_ACC = [], []
    train_batches = len(train_loader)

    VAL_LOSS, VAL_ACC = [], []
    val_batches = len(val_loader)

    # loop for every epoch (training + evaluation)
    start_ts = time.time()
    for e in range(epoch):
        train_losses = 0
        train_accuracies = 0

        # progress bar
        progress = tqdm(enumerate(train_loader), desc="Loss: ", total=train_batches)

        # ----------------- TRAINING  --------------------
        # set model to training
        model.train()

        for i, data in progress:
            X, y = data[0].to(device), data[1].to(device)

            # training step for single batch
            model.zero_grad()

            # forward pass
            outputs = model(X)
            loss = loss_function(outputs, y)

            # backward pass
            loss.backward()
            optimizer.step()

            train_losses += loss.item()

            ps = torch.exp(outputs)
            top_p, top_class = ps.topk(1, dim=1)
            equals = top_class == y.view(*top_class.shape)
            train_accuracies += torch.mean(equals.type(torch.FloatTensor)).item()

            # updating progress bar
            progress.set_description("Loss: {:.4f}".format(train_losses/(i+1)))

        TRAIN_ACC.append(train_accuracies/train_batches)
        TRAIN_LOSS.append(train_losses/train_batches)

        # releasing unceseccary memory in GPU
        if torch.cuda.is_available():
            torch.cuda.empty_cache()

        # ----------------- VALIDATION  -----------------
        val_losses = 0
        val_accuracies = 0

        # set model to evaluating (testing)
        model.eval()
        with torch.no_grad():
            for i, data in enumerate(val_loader):
                X, y = data[0].to(device), data[1].to(device)
                outputs = model(X) # this gives the prediction from the network
                val_losses += loss_function(outputs, y).item()

                ps = torch.exp(outputs)
                top_p, top_class = ps.topk(1, dim=1)
                equals = top_class == y.view(*top_class.shape)
                val_accuracies += torch.mean(equals.type(torch.FloatTensor)).item()

        print("Epoch {}/{} >> Training loss: {:.3f}, Validation loss: {:.3f}, Validation accuracy: {:.3f}".format(
            e+1, epoch, train_losses/train_batches, val_losses/val_batches, val_accuracies/val_batches*100)
        )

        VAL_ACC.append(val_accuracies/val_batches)
        VAL_LOSS.append(val_losses/val_batches)

    tr_time = time.time()-start_ts
    print("Training time: {:.3f}s".format(tr_time))

    return {
    "model" : model.model_selection,  # Mengganti model.name dengan model.model_selection
    "train_acc" : TRAIN_ACC,
    "train_loss" : TRAIN_LOSS,
    "val_acc" : VAL_ACC,
    "val_loss" : VAL_LOSS,
    "exc_time" : tr_time
}
```

* To visualize the model performance

```python
def plot_performance(dict_ : dict, *args, **kwargs) -> None:
    my_figure = plt.figure(figsize=(12, 4))
    # NOTE: figsize=(width/horizontally, height/vertically)

    m = my_figure.add_subplot(121)
    plt.plot(dict_["train_loss"], label="Train Loss")
    plt.plot(dict_["val_loss"], label="Valid. Loss")
    plt.title("LOSS")
    plt.xlabel("Epoch")
    plt.ylabel("Score")
    plt.legend(loc="best")

    n = my_figure.add_subplot(122)
    plt.plot(dict_["train_acc"], label="Train Accuracy")
    plt.plot(dict_["val_acc"], label="Valid. Accuracy")
    plt.title("ACCURACY")
    plt.xlabel("Epoch")
    plt.ylabel("Score")
    plt.legend(loc="best")

    plt.tight_layout()
    plt.show()
```

```python
class VisionModel(torch.nn.Module):
    def __init__(self, model_selection: str, *args, **kwargs):
        super(VisionModel, self).__init__()
        self.model_selection = model_selection

        if self.model_selection == "resnet":
            self.model = torchvision.models.resnet18(pretrained=True)
            # Ganti input layer untuk Grayscale image
            self.model.conv1 = torch.nn.Conv2d(in_channels=1, out_channels=64, kernel_size=7, stride=2, padding=3, bias=False)
            # Ganti output layer untuk 10 kelas (sesuaikan dengan dataset Anda)
            self.model.fc = torch.nn.Linear(self.model.fc.in_features, 10)

        elif self.model_selection == "densenet":
            self.model = torchvision.models.densenet121(pretrained=True)
            # Ganti input layer untuk Grayscale image
            self.model.features.conv0 = torch.nn.Conv2d(in_channels=1, out_channels=64, kernel_size=7, stride=2, padding=3, bias=False)
            # Ganti output layer untuk 10 kelas (sesuaikan dengan dataset Anda)
            self.model.classifier = torch.nn.Linear(self.model.classifier.in_features, 10)

        elif self.model_selection == "vit":
            # Ini hanya contoh, Anda perlu menyesuaikan dengan model Vision Transformer (ViT) yang digunakan
            self.model = torchvision.models.vit_b16(pretrained=True)
            # Ganti input layer untuk Grayscale image
            self.model.patch_embed = torch.nn.Conv2d(in_channels=1, out_channels=self.model.patch_embed.out_channels, kernel_size=3, stride=1, padding=1)

            # Ganti output layer untuk 10 kelas (sesuaikan dengan dataset Anda)
            self.model.head = torch.nn.Linear(self.model.head.in_features, 10)

        self.softmax = torch.nn.Softmax(dim=1)

    def forward(self, data, *args, **kwargs) -> torchvision.models:
        x = self.model(data)
        return self.softmax(x)

# class VisionModel(torch.nn.Module):
#     def __init__(self, model_selection : str, *args, **kwargs) -> None:
#         super(VisionModel, self).__init__()
#         self.model_selection = self.name = model_selection
#         self.in_channels = 1

#         def create_conv2d(this_layer, *args, **kwargs) -> torch.nn.modules.conv.Conv2d:
#             return torch.nn.Conv2d(
#                 in_channels=self.in_channels, out_channels=this_layer.out_channels,
#                 kernel_size=this_layer.kernel_size, stride=this_layer.stride,
#                 padding=this_layer.padding, bias=this_layer.bias
#             )

#         if not self.model_selection.lower() in ["resnet", "densenet", "vit"]:
#             raise ValueError("Please select the model: 'resnet', 'densenet', or 'vit'.")

#             if self.model_selection == "resnet":
#                 self.model = torchvision.models.resnet18(pretrained=True)
#                 self.model.conv1 = create_conv2d(self.model.conv1)
#                 self.model.fc = torch.nn.Linear(self.model.fc.in_features, 10)

#             elif self.model_selection == "densenet":
#                 self.model = torchvision.models.densenet121(pretrained=True)
#                 self.model.features.conv0 = create_conv2d(self.model.features.conv0)
#                 self.model.classifier = torch.nn.Linear(self.model.classifier.in_features, 10)

#             elif self.model_selection == "vit":
#                 self.model = torchvision.models.vit_b_16(pretrained=True)
#                 self.model.conv_proj = create_conv2d(self.model.conv_proj)
#                 self.model.head = torch.nn.Linear(self.model.head.in_features, 10)

#             self.softmax = torch.nn.Softmax(dim=1)

#         def forward(self, data, *args, **kwargs):
#             x = self.model(data)
#             return self.softmax(x)

#     #     if self.model_selection == "resnet":
#     #         self.model = torchvision.models.resnet18(pretrained=True)
#     #         self.model.conv1 = create_conv2d(self.model.conv1) # change the input layer to take Grayscale image, instead of RGB
#     #         self.model.fc = torch.nn.Linear(self.model.fc.in_features, 10) # change the output layer to output 10 classes

#     #     elif self.model_selection == "densenet":
#     #         self.model = torchvision.models.densenet121(pretrained=True)
#     #         self.model.features.conv0 = ... # TODO: Change the DenseNet input layer stack by calling create_conv2d()
#     #         self.model.classifier = ... # TODO: Change the DenseNet output layer with 10 classes

#     #     elif self.model_selection == "vit":
#     #         self.model = torchvision.models.vit_b_16(pretrained=True)
#     #         self.model.conv_proj = ... # TODO: Change the ViT input layer stack by calling create_conv2d()
#     #         self.model.classifier = # TODO: Change the ViT output layer with 10 classes

#     #     self.softmax = torch.nn.Softmax(dim=1)

#     # def forward(self, data, *args, **kwargs) -> torchvision.models:
#     #     x = self.model(data)
#     #     return self.softmax(x)
```

Dalam pelatihan Transfer Learning dengan layer yang dibekukan (frozen layers), hasil akurasi akhir yang lebih rendah dibandingkan dengan pelatihan tanpa pembekuan layer dapat dijelaskan dengan beberapa faktor:

Layer yang Dibekukan: Ketika Anda menggunakan Transfer Learning, Anda biasanya mengambil model yang telah dilatih pada tugas yang berbeda (misalnya, ImageNet) dan kemudian mengadaptasikannya untuk tugas yang Anda kerjakan. Jika Anda memutuskan untuk membekukan beberapa layer selama adaptasi, layer-layer tersebut mungkin tidak dapat menyesuaikan representasi gambar Anda dengan baik. Ini dapat mengakibatkan kehilangan informasi yang penting untuk tugas Anda.

Keterbatasan Data: Jika jumlah data pelatihan yang Anda miliki terbatas, pembekuan layer dapat menghambat model Anda untuk mempelajari representasi yang lebih spesifik untuk tugas Anda. Jika data Anda berbeda secara signifikan dari data yang digunakan untuk melatih model awal, pembekuan layer bisa membatasi kemampuan model untuk beradaptasi.

Jumlah Epoch: Jumlah epoch pelatihan yang Anda tentukan juga dapat mempengaruhi hasil akhir. Dengan jumlah epoch yang sama, model yang tidak memiliki layer yang dibekukan mungkin memiliki lebih banyak kesempatan untuk mengubah parameter dan menyesuaikan representasinya.

Tugas yang Berbeda: Model yang awalnya dilatih pada tugas yang berbeda (seperti ImageNet) mungkin memiliki representasi yang tidak sepenuhnya sesuai dengan tugas yang Anda kerjakan. Pembekuan layer dalam Transfer Learning dapat membuatnya sulit bagi model untuk mengubah representasi ini untuk tugas Anda.

Untuk memperbaiki hasil akhir dengan Transfer Learning dan layer yang dibekukan, Anda mungkin perlu mempertimbangkan beberapa hal:

Coba untuk tidak membekukan terlalu banyak layer atau hanya membebaskan beberapa layer konvolusi terakhir.
Pastikan bahwa data pelatihan yang Anda miliki cukup representatif dan beragam.
Tingkatkan jumlah epoch pelatihan untuk memberikan model lebih banyak waktu untuk beradaptasi.
Penting untuk diingat bahwa tidak ada pendekatan yang benar atau salah, dan hasil pelatihan dapat bervariasi tergantung pada dataset, tugas, dan arsitektur model yang Anda gunakan. Eksperimen dengan berbagai konfigurasi dapat membantu Anda menemukan yang paling sesuai dengan kebutuhan Anda.

Penurunan akurasi pada awal (epok pertama) ketika lebih banyak layer dibekukan dalam pelatihan model dapat dijelaskan oleh beberapa faktor:

Transfer Learning with Frozen Layers: Ketika Anda membekukan lebih banyak layer, Anda menghentikan proses penyesuaian parameter-parameter dalam layer-layer tersebut. Dalam epok pertama, model tidak memiliki kesempatan untuk menyesuaikan representasinya dengan baik terhadap tugas yang sedang Anda kerjakan. Karena model asalnya dilatih untuk tugas yang berbeda, representasi gambar mungkin tidak sesuai dengan baik dengan dataset Anda. Seiring berjalannya waktu dan peningkatan dalam epok, model memiliki kesempatan untuk menyesuaikan representasi ini, dan akurasi akan meningkat.

Informasi Awal yang Terbatas: Saat lebih banyak layer dibekukan, model mungkin kehilangan kemampuan untuk mempelajari informasi yang sangat penting untuk tugas Anda dari gambar. Ini dapat mengakibatkan akurasi awal yang rendah. Pada awalnya, model hanya memiliki akses terhadap informasi yang tersedia dalam layer-layer yang tidak dibekukan, dan ini mungkin tidak mencukupi untuk tugas Anda.

Data yang Berbeda: Jika data pelatihan Anda sangat berbeda dari data yang digunakan dalam pelatihan model awal, maka pembekuan layer-layer dalam model dapat membuatnya sulit untuk beradaptasi dengan baik pada data baru. Ini bisa menjadi faktor kontribusi terhadap penurunan akurasi awal.

Kompleksitas Model: Jika Anda memiliki model yang sangat dalam dan kompleks, maka ketika lebih banyak layer dibekukan, representasi yang dihasilkan oleh layer yang tidak dibekukan mungkin tidak cukup ekspresif untuk mengatasi tugas Anda pada awalnya.
